## 2. 연습 문제 4.1.2
````{admonition} 연습 문제 4.1.2

(1) 열의 갯수와 행의 갯수가 각각 5개 이상이어야 한다.
(2) 열에는 정수, 문자열, 실수 자료형 데이터가 각각 1개 이상씩 포함되어 있어야 한다.
````

`````
d = pd.DataFrame({
    "은학": [83, "수학(가)", 92, 76, 45, 50, 27],
    "지민": [68, "수학(가)", 88, 90, 42, 45, 44],
    "종현": [91, "수학(나)", 96, 94, 45, 45, 50],
    "준호": [96, "수학(나)", 88, 100, 50, 50, 47],
    "종완": [96, "수학(가)", 100, 100, 50, 50, 50]
}, index=["국어", "수학 유형", "수학", "영어", "한국사", "탐구1", "탐구2"])

d.loc["평균"] = pd.concat([d.loc[["국어", "수학", "영어"], :], d.loc[["한국사", "탐구1", "탐구2"], :] * 2]).mean()
d
`````

```

    "국어": [80, 90, 70, 30],
    "영어": [90, 70, 60, 40],
    "수학": [90, 60, 80, 70],

```

(1) 모든 학생의 수학 점수를 시리즈로 나타낸다.
(2) 모든 학생의 국어와 영어 점수를 데이터 프레임으로 나타낸다.
